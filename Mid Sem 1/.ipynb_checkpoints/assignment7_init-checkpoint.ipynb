```python
%matplotlib inline

import pandas as pd;
import numpy as np;
import matplotlib.pyplot as pyplot;
```

```python
data=pd.read_excel("NUMBER_OF_FOREIGN_TOURISTS_TO_INDIA.xls", sheet_name="Foreign Tourists To India");
```

```python
data.fillna(0)
```

```python
#number of tourists to India in each year
print(data.iloc[:87,2:].sum());
```

```python
Nationality_Tourists_desc=data[data['Nationality']!="Total"].sort_values(by='Total', ascending=False);
print(Nationality_Tourists_desc['Nationality'].head(11));
```

```python
Nationality_Tourists_desc_2001=data[data['Nationality']!="Total"].sort_values(by="2001", ascending=False);
print(Nationality_Tourists_desc_2001['Nationality'].head(11));
```

```python
Nationality_Tourists_desc_2010=data[data['Nationality']!="Total"].sort_values(by='2010', ascending=False);
print(Nationality_Tourists_desc_2010['Nationality'].head(11));
```
