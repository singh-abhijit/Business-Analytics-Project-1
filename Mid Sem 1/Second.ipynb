```python
temp2015=temp_data_region.sort_values(by='2015', ascending=False);
temp1=temp2015.iloc[:4];
temp1
```

```python
temp2=temp2015.iloc[4:8]
temp2
```

```python
temp3=temp2015.iloc[8:];
temp3
```

```python
melted_temp1=pd.melt(temp1, id_vars=['Region'], var_name='Year', value_name='FTA');
melted_temp2=pd.melt(temp2, id_vars=['Region'], var_name='Year', value_name='FTA');
melted_temp3=pd.melt(temp3, id_vars=['Region'], var_name='Year', value_name='FTA');
```

```python
melted_temp1.head(7)
```

```python
pyplot.style.use('ggplot');

a1=sns.relplot(x="Year", y="FTA", data=melted_temp1, kind='line', style='Region', hue='Region', markers=True, dashes=False);
a1.fig.suptitle("FTA to India 2001-2015", y=1.05)
a1.set(xlabel='Years', ylabel='FTA');
# sns.set_context('notebook');
pyplot.xticks(rotation=450);
pyplot.show();
pyplot.clf();
```
